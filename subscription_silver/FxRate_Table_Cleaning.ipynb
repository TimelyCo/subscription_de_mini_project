```python
bronze_schemapath  = "`bronze_catalog`.default"
silver_schemapath = "`silver_catalog`.default"
```

```python
tables = spark.catalog.listTables(dbName=bronze_schemapath)
```

```python
fivetran_ingested_column_to_remove = [
    "_file", 	
    "_fivetran_synced", 
    "_modified",
    "_line"
]
```

```python
def remove_fivetran_ingested_column(df):
    for column in fivetran_ingested_column_to_remove:
        df = df.drop(column)
    return df
```

```python
# Read from bronze, remove metadata columns, and write to silver
df = spark.read.table(f"{bronze_schemapath}.fixrate_table")
df_clean = remove_fivetran_ingested_column(df)
display(df_clean)
#df_clean.write.format("delta").mode("overwrite").saveAsTable(f"{silver_schemapath}.fixrate_table")
```

```python
from pyspark.sql.functions import to_date, col

df = spark.read.table(f"{silver_schemapath}.fixrate_table")
df = df.withColumn("effective_date", to_date(col("effective_date"), "dd-MM-yyyy"))
#df.write.format("delta").mode("overwrite").saveAsTable(f"{silver_schemapath}.fixrate_table")
```
