```python
azure_ingested_schemapath = "`subscription_mini_project`.azure_blob_storage"
bronze_schemapath  = "`bronze_catalog`.default"
```

```python
tables = spark.catalog.listTables(bronze_schemapath)
```

```python
fivetran_ingested_column_to_remove = [
    "_file", 	
    "_fivetran_synced", 
    "_modified",
    "_line"
]
```

```python
def remove_fivetran_ingested_column(df):
    for column in fivetran_ingested_column_to_remove:
        df = df.drop(column)
    return df
```

```python
for table in tables:
    df = spark.read.table(f"{bronze_schemapath}.{table.name}")
    df = remove_fivetran_ingested_column(df)
    df.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable(f"{bronze_schemapath}.{table.name}")
```
